```python
# Explore which vaccine taken by which country
df_vac_coutary = df[['vaccines','country']]
df_vac_coutary.groupby('country').head()
```

```python
df_vac_coutary.vaccines.unique()
```

Now let’s see how many countries are taking each of the vaccines mentioned in this data:

```python
dict = {}

for i in df_vac_coutary.vaccines.unique():
    dict[i] = [df_vac_coutary["country"][j] for j in df_vac_coutary[df_vac_coutary["vaccines"]==i].index]
    
vaccines = {}

for key, value in dict.items():
    vaccines[key] = set(value) # set gives unique value
    
for i, j in vaccines.items():
    print(f"{i}:::{j}")       
    
```
